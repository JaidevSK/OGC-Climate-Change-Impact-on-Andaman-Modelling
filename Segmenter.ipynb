```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
original = cv2.imread('BioDiversityMap.png')
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB) 
original.shape
```

```python
plt.figure(figsize = (20, 10))
plt.imshow(original)
```

```python
original = cv2.resize(original, (2600, 11100))
```

```python
original.shape
```

```python
plt.imshow(original)
```

```python
# Convert the image into HSV
hsv_image = cv2.cvtColor(original, cv2.COLOR_RGB2HSV)

lower = np.array([0, 0, 224])
upper = np.array([0, 0, 255])

mask = cv2.inRange(hsv_image, lower, upper)

land = cv2.bitwise_and(original, original, mask=mask)
land = ~land
```

```python
plt.figsize = (20, 20)
plt.imshow(land)
```

```python
lightgreen = cv2.imread('LightGreen.png')
hsv_lg = cv2.cvtColor(lightgreen, cv2.COLOR_BGR2HSV)
print(hsv_lg)

hsv_image = cv2.cvtColor(original, cv2.COLOR_RGB2HSV)

lower = np.array([40, 150, 150])
upper = np.array([60, 255, 255])

mask = cv2.inRange(hsv_image, lower, upper)

grassland = cv2.bitwise_and(original, original, mask=mask)

plt.figsize = (20, 20)
plt.imshow(grassland)
cv2.imwrite('grassland.png', grassland)
```

```python
Grassland_segmented = cv2.imread('grassland.png')
Grassland_segmented = cv2.cvtColor(Grassland_segmented, cv2.COLOR_BGR2GRAY)
Grassland_segmented = (Grassland_segmented>0)*255
cv2.imwrite('Grassland_segmented.png', Grassland_segmented)
plt.imshow(Grassland_segmented)
```
